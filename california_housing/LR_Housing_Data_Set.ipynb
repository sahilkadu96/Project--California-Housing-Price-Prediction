```python
from sklearn.datasets import  fetch_california_housing
```

```python
df=fetch_california_housing()
print(df.DESCR)
```

```python
import pandas as pd

data=pd.DataFrame(df.data,columns=df.feature_names)
data
```

```python
data.shape
```

```python
data.columns

```

```python
x=data
y=df.target
```

```python
x
```

```python
y.shape
```

```python
y
```

```python
x.describe()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.25, random_state=0)

SC=StandardScaler()
LR =LinearRegression()

x_train_scaled=SC.fit_transform(x_train)

LR.fit(x_train,y_train)
pred=LR.predict(x_test)

print(r2_score(pred,y_test))
print(mean_squared_error(pred,y_test))
print(mean_absolute_error(pred,y_test))
```

```python
import pickle
pickle.dump(LR,open('model.pkl','wb'))

```

```python
import joblib
import matplotlib.pyplot as plt
import pandas as pd
import streamlit as st


model = joblib.load('model.pkl')
def web_app():
       st.write("""
       # House Price Predictor Web App
       ## This app predicts the price of houses
        """)
       st.header("User Details")
       st.subheader("Kindly Enter The following Details in order to make a prediction")
       MedInc = st.number_input("Insurance Premium",0,10)
       HouseAge = st.number_input("House Age",0,10)
       AveRooms = st.number_input("Avg Rooms",0,10)
       AveBedrms = st.number_input("Average Bedrooms",0,10)
       Population = st.number_input("Population",0,10)
       AveOccup = st.number_input("Avg Occupation",0,10)
       Latitude = st.number_input("Latitude",0,10)
       Longitude = st.number_input("Longitude",0,10)

       result = model.predict([[MedInc,HouseAge,AveRooms,AveBedrms, Population, AveOccup, Latitude, Longitude]])

       st.text_area(label='The price is:- ',value=result , height= 100)

       if st.button("Press here to make Prediction"):
          run = web_app()
```
